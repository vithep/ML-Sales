Foi criado uma função para contar quantos valores, de cada variável, temos em cada coluna do tipo 'object'.

```python
def obj_count(dataframe):
        
    for i in dataframe.columns:
        
        if dataframe[i].dtypes == 'object':
            
            print('--------------------------',i,'----------------------------------')
        
            print(dataframe[i].value_counts())
        
            print('-----------------------------------------------------------------')
            
```

```python
def obj_pcnt(dataframe):
        
    for i in dataframe.columns:
        
        if dataframe[i].dtypes == 'object':
            
            print('--------------------------',i,'----------------------------------')
        
            print(dataframe[i].value_counts(normalize=True))
        
            print('-----------------------------------------------------------------')
```

Realizando uma Clusterização com o 'LIMITE_CEDENTE' e a 'TAXA_MENSAL'

```python
X = df1[['LIMITE_CEDENTE','TAXA_MENSAL']].iloc[: , :].values
```

```python
cluster_minimo = 1
cluster_maximo = 7
wcss = []
for i in range (cluster_minimo,cluster_maximo):
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(cluster_minimo,cluster_maximo),wcss)
plt.scatter(range(cluster_minimo,cluster_maximo),wcss)
plt.ticklabel_format(style='plain', axis='y')
plt.figure(1 , figsize = (15 ,6))

plt.show()
```

Podemos observar que iremos utilizar 2 Clusters

```python
kmeans = KMeans(n_clusters =2, init='k-means++',max_iter=300,n_init=10,random_state=0)
pred_y = kmeans.fit_predict(X)
plt.scatter(X[:,0],X[:,1],c=pred_y)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=50,c='blue')
plt.ticklabel_format(style='plain', axis='y')
plt.ticklabel_format(style='plain', axis='x')
plt.figure(figsize=(20,10))
plt.show()
```

#### > Random Undersampling

```python
print('Original dataset shape %s' %Counter(y))

rus = RandomUnderSampler(random_state=42)
X_rus,y_rus = rus.fit_resample(X,y)

print('Resampled dataset shape %s' %Counter(y_rus))
```

```python
model = XGBClassifier()
model.fit(X_rus,y_rus)
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print('Accuracy: %.2f%%' %(accuracy*100))
```

```python
conf_mat = confusion_matrix(y_true=y_test, y_pred=y_pred)
print('Confusion matrix:\n', conf_mat)
labels =  ['ATRASO','EM_DIA']
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(conf_mat, cmap=plt.cm.Blues)
fig.colorbar(cax)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)
plt.xlabel('Predicted')
plt.ylabel('Expected')
plt.show()
```

No Undersampling, tivemos uma acurácia pior do que a inicial.

#### > Random Oversampling

```python
print('Original dataset shape %s' %Counter(y))

ros = RandomOverSampler(random_state=42)
X_ros,y_ros = ros.fit_resample(X,y)

print(X_ros.shape[0] - X.shape[0], ' new random picked points')
print('Resampled dataset shape %s' %Counter(y_ros))
```

```python
model = XGBClassifier()
model.fit(X_ros,y_ros)
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print('Accuracy: %.2f%%' %(accuracy*100))
```

```python
conf_mat = confusion_matrix(y_true=y_test, y_pred=y_pred)
print('Confusion matrix:\n', conf_mat)
labels = ['ATRASO','EM_DIA']
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(conf_mat, cmap=plt.cm.Blues)
fig.colorbar(cax)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)
plt.xlabel('Predicted')
plt.ylabel('Expected')
plt.show()
```

No modelo Oversampling, conseguimos uma acurácia melhor, onde obtivemos um valor igual a 92.74%, apresentando acertos iguais a 2620, e erros iguais a 205. Com isso iremos utilizar esse modelo.
